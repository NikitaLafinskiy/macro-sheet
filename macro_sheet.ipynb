```python
def process_func(a):
  if(type(a) != str):
    return f"{round(a * 100)}%"
  elif(type(a) == str):
    return a
```

```python
def process_func_negative(a):
  if(a not in countries.to_numpy()):
    a = int(str(a).split('%')[0])
    a = 100 - a
    return f'{a}%'
  elif(a in countries.to_numpy()):
    return a
```

```python
df_sorted_arr_scaled_100 = []
for column in range(25):
  regularized_df = df_sorted_arr_scaled[column].applymap(process_func)
  for column in regularized_df:
    if(column not in positive_num_cols):
      regularized_df[column] = regularized_df[column].apply(process_func_negative)
      regularized_df.sort_values(by=column, ascending=False)
    df_sorted_arr_scaled_100.append(regularized_df)
    # regularized_df.to_csv(f'./regularized1/By_{column}.csv')
```

```python
countries = countries.to_numpy()
```

```python
wealth_weights = [0.3, 0.1, 0.3, 0.2, 0.1]
healthcare_weights = [0.3, 0.2, 0.2, 0.1, 0.2]
safety_weights = [0.25, 0.3, 0.15, 0.2, 0.1]
education_weights = [0.25, 0.2, 0.3, 0.15, 0.1]
govn_weights = [0.3, 0.25, 0.2, 0.15, 0.1]
```

```python
def calc_val(df, weights=[0.3, 0.2, 0.2, 0.1, 0.2]):
  weighted_country_cols = []
  for country in countries:
    country = df[df['Country'] == country]
    country_weighted_vals = []
    country_vals = country.to_numpy()[0, 1:]
    for i, num in enumerate(country_vals):
      num = int(num.split("%")[0])
      country_weighted_vals.append(num * weights[i])
    weighted_country_cols.append({f"{country['Country'].to_numpy()[0]}":float(np.sum(country_weighted_vals))})
  return weighted_country_cols
```

```python
def convert_dict(d):
  countries = []
  vals = []
  for i in d:
    c = [i for i in i.keys()][0]
    v = [i for i in i.values()][0]
    countries.append(c)
    vals.append(v)
  return pd.DataFrame(np.transpose([countries, [int(i) for i in vals]]), columns=['Country', 'Value']).sort_values(by='Value', ascending=False)
```

```python
wealth_vals = calc_val(wealth, wealth_weights)
healthcare_vals = calc_val(healthcare, healthcare_weights)
edu_vals = calc_val(education, education_weights)
safety_vals = calc_val(safety, safety_weights)
govn_vals = calc_val(govn, govn_weights)
```

```python
healthcare_vals
```

```python
wealth_df = convert_dict(wealth_vals)
health_df = convert_dict(healthcare_vals)
safety_df = convert_dict(safety_vals)
education_df = convert_dict(edu_vals)
govn_df = convert_dict(govn_vals)
```
